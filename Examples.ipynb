# SNaCK examples

SNaCK is an algorithm that generates *concept embeddings* of your data.

As input, give it:

- A d-dimensional feature representation of your data, as a matrix with shape (N, d) and type 'double';
- A set of "triplet constraints," as a matrix with shape (t, 3) and type 'long'

SNaCK will output an embedding that tries to satisfy the neighbor relationships of your data in feature space (like t-SNE), as well as the triplet constraints.

Each row of the triplet matrix represents a single "triplet constraint." A row like (3, 5, 8) specifies that in the final embedding, `X[3]` should be closer to `X[5]` than `X[3]` is to `X[8]`.

Using triplet constraints allows you to capture properties of your data that you can't quite model directly, such as subjective measurements of human taste.

# Method signature

```python
snack.snack_embed(
        # Important:
        double[:, ::1] X_np,
        double contrib_cost_tsne,
        long[:,::1] triplets,
        double contrib_cost_triplets,
        double perplexity = 30.0,
        double theta = 0,

        # Fine-grained optimization control:
        int no_dims = 2,
        alpha = None,
        int max_iter = 300,     # BH-tSNE default: 1000
        early_exaggeration = 4, # BH-tSNE default: 12
        early_exaggeration_switch_iter = 100,  # BH-tSNE default: 250 (see max_iter!)
        double momentum = 0.5,
        int momentum_switch_iter = 250,
        double final_momentum = 0.8,
        double learning_rate = 1.0,
        initial_Y = None,

        # Other:
        each_fun = None,
        verbose = True,
        num_threads = None,
)
```

Returns an array with shape (len(X_np), no_dims). The i-th
row in this array corresponds to the no_dims-dimensional
coordinate of the point.

Important parameters
--------------------

- **`X_np`** : numpy array with shape `(N, D)` and type double.
  - The feature representation of N points in D dimensions.
- **`contrib_cost_tsne`** : double
  - Trade-off: raises the influence of t-SNE in the final embedding. Typical ranges are between 100 and 5000. Suggestion: Pick contrib_cost_tsne and contrib_cost_triplets such that the norms of the corresponding gradients are roughly equal.
- **`triplets`** : numpy array with shape `(T, 3)` and type long.
  - Each row is (a,b,c), which indicates that in the desired embedding Y, then object a should be closer to b than a is to C, or $$|Y_a-Y_b| < |Y_a-Y_c|$$.
- **`contrib_cost_triplets`** : double
  - Trade-off: raises the influence of triplets in the final embedding. Typical ranges are between 0.1 and 5 Suggestion: Pick contrib_cost_tsne and contrib_cost_triplets such that the norms of the corresponding gradients are roughly equal.
- **`perplexity`** : double
  - t-SNE perplexity parameter, controlling the number of expected neighbors for each point. Generally between 10 and 300.
- **`theta`** : double
  - Optimization parameter: set to 0.0 for an exact solution. Higher values are faster and sloppier. Sets of points whose width is smaller than this angle will be collapsed. Try starting with theta=0.5 and adjust as needed.

Optimization parameters
-----------------------

- **`no_dims`** : int
  - Number of dimensions of the final embedding. High-dimensional embeddings are much easier to satisfy (lower training error), but may capture less information. (UNTESTED for >2 dimensions)
- **`alpha`** : int
  - Degrees of freedom in student T kernel for t-STE. Default is no_dims-1. Considered "black magic"; roughly, how much of an impact badly satisfying points have on the gradient calculation.
- **`max_iter`** : int
  - Number of iterations to convergence. 300 or 500 is usually enough.
- **`early_exaggeration`** : double
  - Magic t-SNE parameter. (see paper)
- **`early_exaggeration_switch_iter`** : int
  - Which iteration to stop using early exaggeration.
- **`momentum`** : double
  - Magic optimization.
- **`final_momentum`** : double
- **`momentum_switch_iter`** : int
  - After momentum_switch_iter iterations, switch momentum to the value of final_momentum.
- **`learning_rate`** : double
  - All gradients are multiplied by this value. initial_y : numpy array with shape (N, no_dims) and type double. Matrix with shape (N, no_dims). If not specified, the embedding will be calculated completely randomly. If unspecified, embedding will be created randomly.

Other parameters
----------------

- **`each_fun`** : function
  - Callback called on each iteration. Accepts parameters: (iteration number, current_Y, C_tSTE, C_tSNE, dY_tSTE, dY_tSNE) Note current_Y is modified each iteration. C_tSTE() and C_tSNE() are thunks that calculate tSTE and tSNE error when called. (This may be expensive!) dY_tSTE and dY_tSNE are the gradients of tSTE and tSNE.
- **`verbose`** : boolean
  - Whether to log debug information. Note that BH-tSNE also does its own logging to stderr from C++.
- **`num_threads`** : int
  - Number of threads to use for OpenMP (but this doesn't really matter much anymore; only the t-STE calculation is parallelized)

# Examples: separating MNIST into Prime and Composite Numbers

In this example, suppose the expert wishes to capture the concept of primality by partitioning the dataset into prime numbers $\{2,3,5,7\}$, composite numbers $\{4,6,8,9\}$ and $\{0,1\}$. Also, for the purpose of this simple example, assume that rather than label the digits directly, the expert compares images based on concept similarity, *i.e.*, primes are more similar to primes than to other images.

We will run SNaCK on the following data as input:
  - Raw pixel intensities;
  - Triplet constraints of the form $(a,b,c)$, where $a$ and $b$ come from the same primality group and $c$ does not

```python
import snack
import time

#Load digits 
digits = load_digits() #fetch_mldata("DIGITS original")
X = digits.data.copy().astype('float')
labels = digits.target.astype('int')

# X = X[::10].copy()
# labels = labels[::10].copy()
#X = GaussianRandomProjection(n_components=70).fit_transform(X)
#labels = labels.copy()

cm = plt.get_cmap('Paired')
colors = np.array([mpl.colors.rgb2hex((cm(c))) for c in np.linspace(0,1,10)])

label_groups = np.arange(10)
label_groups[[0, 1]] = 0
label_groups[[2, 3, 5, 7]] = 1
label_groups[[4, 6, 8, 9]] = 2
index_groups = label_groups[labels]

triplets = []
for i in xrange(1000):
    # Sample a random triplet
    a = np.random.choice(len(X))
    a_group = label_groups[labels[a]]
    # Pick others from within the same group and outside the same group
    b = np.random.choice(np.where(index_groups == a_group)[0])
    c = np.random.choice(np.where(index_groups != a_group)[0])
    triplets.append( (a,b,c) )

triplets = np.vstack(triplets)
```

```python
Y = snack.snack_embed(
    X, 500.0,
    triplets, 0.1,
    theta = 0.5,
    verbose = False,
)
fig,ax = subplots(figsize=(8,8))
ax.set_title("SNACK embedding of digits dataset")
ax.scatter(*Y.T, s=20, lw=0, c=colors[labels])
ax.legend(handles=[matplotlib.patches.Patch(color=colors[i], label="%d"%i) for i in xrange(10)])
```

```python
# Let's explore the effect of many parameters.
C_TRIPLETS = [0.0, 0.1, 0.5, 1.0]
C_TSNE = [0, 100.0, 500.0, 1000]
fig,axes = subplots(len(C_TRIPLETS), len(C_TSNE), figsize=(10, 10))
for c_triplets, row in zip(C_TRIPLETS, axes):
    for c_tsne, ax in zip(C_TSNE, row):
        Y = snack.snack_embed(
            X,        c_tsne,
            triplets, c_triplets,
            theta = 1.5,
            verbose = False,
        )
        ax.scatter(*Y.T, s=20, lw=0, c=colors[labels])

        
```

```python
%%time
def each_fun(iter, Y_np, *other):
    if iter%5==0:
        fig,ax = subplots()
        ax.scatter(*Y_np.T, s=1, c=colors[labels], lw=0)
        from IPython.kernel.zmq.pylab.backend_inline import flush_figures
        flush_figures()
Y = snack.snack_embed(
    X, 500.0,
    triplets, 0.1,
    theta = 0.5,
    each_fun = each_fun,
    verbose = False,
)
scatter(*Y.T, s=2, lw=0, c=labels, cmap='rainbow')
```
